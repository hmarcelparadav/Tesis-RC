# Promedio/Máximo vs DiametroAltura

```python
x = [DiametroAltura90x210, DiametroAltura67x104, DiametroAltura76x128, DiametroAltura38x192]
y = [PromMax90x210, PromMax67x104, PromMax76x128, PromMax38x192 ]

print(x)
print(y)

plt.figure(figsize=(16,9))
plt.ylabel("Prom./Max.", fontweight='bold')
plt.xlabel("Diametro/Altura" , fontweight='bold')
plt.plot(DiametroAltura90x210,PromMax90x210, "^r", ms=10, label = " H=90,D=210 ")
plt.plot(DiametroAltura67x104,PromMax67x104, "ob", ms=10, label = " H=67,D=104 ")
plt.plot(DiametroAltura76x128,PromMax76x128, "sm", ms=10, label = " H=76,D=128 ")
plt.plot(DiametroAltura38x192,PromMax38x192, "pg", ms=10, label = " H=38,D=192 ") #Figura Pentagono
plt.yticks(fontsize=14)
plt.xticks(fontsize=14)
plt.legend()
plt.grid()
plt.show()
```
